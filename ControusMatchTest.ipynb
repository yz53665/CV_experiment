```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
src = cv.imread('ExpPic/car/473.bmp')
template = cv.imread('template.png')
srcGray = cv.cvtColor(src, cv.COLOR_BGR2GRAY)
templateGray = cv.cvtColor(template, cv.COLOR_BGR2GRAY)
```

```python
thr, srcOSTU = cv.threshold(srcGray, 0, 255, cv.THRESH_OTSU)
thr, templateOSTU = cv.threshold(templateGray, 0, 255, cv.THRESH_OTSU)
```

```python
plt.subplot(121)
plt.imshow(srcOSTU, cmap='gray')
plt.subplot(122)
plt.imshow(templateOSTU, cmap='gray')
```

```python
cannyTemplate = cv.Canny(template, 40, 100)
cannySrc = cv.Canny(src, 50, 150)
plt.figure(figsize=(8,8), dpi=100)
plt.subplot(121)
plt.imshow(cannyTemplate, cmap='gray')
plt.subplot(122)
plt.imshow(cannySrc, cmap='gray')
```
